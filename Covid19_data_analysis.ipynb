```python
df=data.drop(['Lat','Long'],axis=1,inplace=True)
```

```python
countries=list(corona_dataset_aggrated.index)
max_infection_rates=[]
for c in countries:
    max_infection_rates.append(corona_dataset_aggrated.loc[c].diff().max())
corona_dataset_aggrated["Max_infection_rates"]=max_infection_rates
```

```python
corona_dataset_aggrated
```

```python
corona_data=pd.DataFrame(corona_dataset_aggrated['Max_infection_rates'])
```

```python
corona_data
```

```python
happiness_report=pd.read_csv("worldwide_happiness_report.csv")
happiness_report
```

```python
happiness_report.drop(['Overall rank','Score','Generosity','Perceptions of corruption'],axis=1,inplace=True)
```

```python
happiness_report.head()
```

```python
happiness_report.set_index("Country or region",inplace=True)
happiness_report
```

```python
happiness_report.shape
```

```python
data=corona_data.join(happiness_report,how="inner")
data
```
